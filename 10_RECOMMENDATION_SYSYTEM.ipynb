```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
book=pd.read_csv('book.csv',encoding='Latin1')
book
```

```python
books1=book.drop(['Unnamed: 0'],axis=1)
books1
```

```python
books2=books1.rename({'User.ID':'userid','Book.Title':'bookstitle','Book.Rating':'rating'},axis=1)
books2
```

```python
# Sort by User IDs
books2.sort_values(['userid'])
```

```python
# number of unique users in the dataset
len(books2['userid'].unique())
```

```python
# number of unique books in the dataset
len(books2['bookstitle'].unique())
```

```python
# converting long data into wide data using pivot table
books3=books2.pivot_table(index='userid',columns='bookstitle',values='rating').reset_index(drop=True)
books3
```

```python
books3.index=books2['userid'].unique()
books3
```

```python
# Impute those NaNs with 0 values
books3.fillna(0,inplace=True)
books3
```

```python
# Calculating Cosine Similarity between Users on array data
user_sim=1-pairwise_distances(books3.values,metric='cosine')
user_sim
```

```python
# Store the results in a dataframe format
user_sim2=pd.DataFrame(user_sim)
user_sim2
```

```python
# Set the index and column names to user ids 
user_sim2.index=books2['userid'].unique()
user_sim2.columns=books2['userid'].unique()
user_sim2
```

```python
# Nullifying diagonal values
np.fill_diagonal(user_sim,0)
user_sim2
```

```python
user_sim2.idxmax(axis=1)
```
